```python
meals = [
    {"name": "Moong Dal", "calories": 350, "protein": 24, "carbs": 60, "fat": 1, "fiber": 15, "serving_grams": 100},
    {"name": "Soya Chunks", "calories": 350, "protein": 50, "carbs": 30, "fat": 1, "fiber": 12, "serving_grams": 100},
    {"name": "Vada Pav", "calories": 290, "protein": 6, "carbs": 45, "fat": 10, "fiber": 5, "serving_grams": 100},
    {"name": "Scrambled Eggs", "calories": 400, "protein": 36, "carbs": 3, "fat": 30, "fiber": 0, "serving_grams": 300},
    {"name": "Chicken Breast", "calories": 350, "protein": 70, "carbs": 0, "fat": 8, "fiber": 0, "serving_grams": 300},
    {"name": "Basa Fish", "calories": 250, "protein": 50, "carbs": 0, "fat": 7, "fiber": 0, "serving_grams": 300},
    {"name": "Bajra Bhakari", "calories": 350, "protein": 10, "carbs": 70, "fat": 5, "fiber": 10, "serving_grams": 100},
    {"name": "Cooked Rice", "calories": 120, "protein": 3, "carbs": 30, "fat": 0.5, "fiber": 0.5, "serving_grams": 100}
]

for meal in meals:
    print(meal["name"], "-", meal["calories"], "calories |", meal["protein"], "g protein |", meal["carbs"], "g carbs |", meal["fat"], "g fat |", meal["fiber"], "g fiber |", meal["serving_grams"], "grams")
```

```python
highest_calorie_meal = max(meals, key= lambda meal: meal["calories"])
print("Highest_calorie_meal:", highest_calorie_meal["name"], "-", highest_calorie_meal["calories"], "calories")
```

```python
total_protein = sum(meal["protein"] for meal in meals)
print("Total protein:", total_protein, "g")
```

```python
lowest_calorie_meal = min(meals, key= lambda meal: meal["calories"])
print("Lowest_calorie_meal:", lowest_calorie_meal["name"], "-", lowest_calorie_meal["calories"], "calories")
```

```python
average_calories = sum(meal["calories"] for meal in meals) / len(meals)
print("Average calories:", average_calories)
```
